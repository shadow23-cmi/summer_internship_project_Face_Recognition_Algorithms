### Importing necessary libraries

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pickle
import random
```

### Different directories

```python
dir_nonface="/content/nonface_resized/"
dir_face_ufi_cropped="/content/sd18_24_24/"
dir_face_mugshot="/content/mugshot_24_24/"
dir_dumping="/content/cifar-10-python/"
```

### Creating the list for pickling

```python
#Creating different lists for face and nonface with proper classification
face_sample=list(zip([cv2.imread(dir_face_ufi_cropped+i,0) for i in os.listdir(dir_face_ufi_cropped)[-50:]],[1 for i in range(50)]))
non_face_sample=list(zip([cv2.imread(dir_nonface+i,0) for i in os.listdir(dir_nonface)[1000:1050]],[0 for i in range(50)]))

#joining the two lists
sample=face_sample+non_face_sample

#shuffing the items of the list
np.random.shuffle(sample)

```

### Check if the list created to be made into into pickle file has proper classification

```python
for i in range(len(sample)):
    print("Image classification for \nthe image below :",sample[i][1])
    plt.imshow(sample[i][0],cmap="gray")
    plt.show()
```

### Creating the pickle file

```python
pickle.dump(sample,open(dir_dumping+"test_50_50.pickle","wb"))
```
